```python
import numpy as np
import tensorflow as tf
import cv2
import pytesseract
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Mount Google Drive (if needed)
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

# Load the image from Google Drive
image_path = '/content/drive/MyDrive/CV_Image/cv3.png'
image = Image.open(image_path)

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Perform OCR
text = pytesseract.image_to_string(image)
print(text)
```

```python
import re

# Preprocess the extracted text
def preprocess_text(text):
    # Remove non-alphanumeric characters and extra whitespaces
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    
    # Normalize the text by converting it to lowercase
    text = text.lower()
    
    return text

# Preprocess the extracted text
preprocessed_text = preprocess_text(text)
print(preprocessed_text)
```

```python
def load_keywords_from_file(file_path):
    with open(file_path, 'r') as file:
        keywords = file.read().splitlines()
    return keywords

def extract_experiences(cv_text, experiences_keywords):
    experiences = []
    # Iterate over each experience keyword
    for keyword in experiences_keywords:
        if keyword in cv_text:
            experiences.append(keyword)
    return experiences

def extract_educations(cv_text, educations_keywords):
    educations = []
    # Iterate over each education keyword
    for keyword in educations_keywords:
        if keyword in cv_text:
            educations.append(keyword)
    return educations

def extract_location(cv_text, location_keywords):
    location = []
    # Iterate over each education keyword
    for keyword in location_keywords:
        if keyword in cv_text:
            location.append(keyword)
    return location

def extract_major(cv_text, major_keywords):
    major = []
    # Iterate over each education keyword
    for keyword in major_keywords:
        if keyword in cv_text:
            major.append(keyword)
    return major

def extract_skills(cv_text, skills_keywords):
    skills = []
    # Iterate over each skill keyword
    for keyword in skills_keywords:
        if keyword in cv_text:
            skills.append(keyword)
    return skills

# Load skill, experience, and education keywords from file
skills_keywords = load_keywords_from_file('/content/skills.txt')
experiences_keywords = load_keywords_from_file('/content/experiences.txt')
educations_keywords = load_keywords_from_file('/content/educations.txt')
major_keywords = load_keywords_from_file('/content/major.txt')
location_keywords = load_keywords_from_file('/content/location.txt')

# Extract skills, experience, and educations
skills = extract_skills(preprocessed_text, skills_keywords)
experience = extract_experiences(preprocessed_text, experiences_keywords)
educations = extract_educations(preprocessed_text, educations_keywords)
major = extract_major(preprocessed_text, major_keywords)
location = extract_location(preprocessed_text, location_keywords)
```

```python
# Print the extracted skills, experience, and educations
print("Experiences:", experience)
print("Educations:", educations)
print("Location:", location)
print("Major:", major)
print("Skills:", skills)
```
